```python
def list_sorting(lst):
    
    """ (list) -> (list)
    
    Return the kind of values in the list.
    Delete the repetitive values and return the list which has no repetitive values.
    
    >>> a = [1, 1, 2, 3, 4, 4, 5 ]
    >>> list_sorting(a)
    [1, 2, 3, 4, 5]
    """
    
    sorting_list = lst[:]
    sorting_list.sort()
    temp = [] # 중복된 값의 인덱스 값이 저장될 임시 리스트 변수

    for i in range(len(sorting_list) - 1):
        if (sorting_list[i] == sorting_list[i+1]):
            temp.append(i)       

    temp.reverse()

    for i in temp:
        del sorting_list[i] 
        
    return sorting_list

def getYear(year):
    
    """ (str) -> (str)
    
    Return the year in the given date.
    
    >>> str1 = "08:54:06 PM 08/23/2018"
    >>> getYear(str1)
    2018
    """
    
    length = len(year)
    result = year[length - 4] + year[length - 3] + year[length - 2] + year[length - 1]
    return result

def getmaximum(lst):
    
    """(list) -> (int)
    
    Return the maximum value in the list.
    
    >>> a = [1, 9, 5, 4, 17,6]
    >>> getmaximum(a)
    17
    """
    
    maximum = lst[0]
    for i in range(len(lst)):
        if maximum < lst[i]:
            maximum = lst[i]
        
    return maximum
```
